# how to filter data

```python
# fetch records related to male passenger
data.loc[data['Sex']=='male']
```

```python
data.loc[data['Sex']=="female"]
```

```python
# how many passenger age is greater than 60?
data.loc[data["Age"]>60]
```

```python
# count
data.loc[data["Age"]>60].shape
```

```python
# less than age 10?
data.loc[data["Age"]<10]
```

```python
# how many passenger less than 10
data.loc[data["Age"]<10].shape
```

```python
# how many female passenger are travelling with an age greater than 60
data.loc[(data["Sex"]=="female")&(data["Age"]>60)].shape
```

```python
# how many female passenger survived
data.loc[(data["Sex"]=="female")&(data['Survived']==1)].shape
```

```python
# how many male passenger died
data.loc[(data["Sex"]=="male")&(data['Survived']==0)].shape

```

```python
# Find the number of passenger with an age less than 10 or greater than 70
data.loc[(data["Age"]<10)|(data["Age"]>70)].shape
```

# type casting
