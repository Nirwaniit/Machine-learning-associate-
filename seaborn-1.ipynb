```python
# import basic libraries
import numpy as np
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
```

```python
# seaborn has a inbuilt dataset
sns.get_dataset_names()
```

```python
# load iris data
data=sns.load_dataset("iris")
```

```python
data
```

# Relation plot
#### relation plots are used to understand relationship between two variables
   * line plot
   * scatter
   * relplot

```python


sns.lineplot(data=data,x="sepal_length",y="sepal_width",color="green") 
```

```python
plt .figure(figsize=(3,3))
sns.lineplot(data=data,x="sepal_length",y="sepal_width",color="green",ci=None)
```

```python
plt .figure(figsize=(6,3))
sns.lineplot(data=data,x="sepal_length",y="sepal_width",ci=None,hue="species",palette="rainbow")  # palette for colors
```

# scatter plot

```python
plt.figure(figsize=(5,3))
sns.scatterplot(data=data,x="petal_length",y="petal_width",hue="species")
```

```python
plt.figure(figsize=(5,3))
sns.scatterplot(data=data,x="petal_length",y="petal_width",hue="species",palette='rainbow')
```

#  rel plot

```python
plt.figure(figsize=(5,3))
sns.relplot(data=data,x="petal_length",y="petal_width")
```

```python
plt.figure(figsize=(5,3))
sns.relplot(data=data,x="petal_length",y="petal_width",kind="line")
```
