# categorical plots
#### it is used to analyze one numerical and one categorical column
- count plot
- box plot
- bar plot
- violin plot
- strip plot
- swarm plot
- cat plot

```python
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
import matplotlib.pyplot as plt
```

```python
# seaborn has a inbuilt dataset
sns.get_dataset_names()
```

```python
# load iris data
data=sns.load_dataset("iris")
```

```python
data
```

```python
data.species.unique()
```

```python
# count plot is used represent the count of categories in categorical column
# count plot requires one categorical column
```

```python
plt.figure(figsize=(5,2))
sns.countplot(data=data,x="species")
```

### bar plot is used to analyse categorical column with numerical column

```python
#  find which species has highest sepal length
plt.figure(figsize=(5,2))
sns.barplot(data=data,x="species",y="sepal_length")
```

```python
# box plot
plt.figure(figsize=(4,2))
sns.boxplot(data=data,x="petal_length")
```

```python
plt.figure(figsize=(4,4))
sns.boxplot(data=data,x="species",y="petal_length")
```

```python
# viol plot
plt.figure(figsize=(4,4))
sns.violinplot(data=data,x="species",y="petal_length")
```

```python
sns.stripplot(data=data,x="species",y="sepal_length", palette="rainbow")
```

```python
sns.swarmplot(data=data,x="species",y="sepal_length", palette="rainbow")
```

```python
# cat plot: using cat plot we can get all the graphs
sns.catplot(data=data,x="species",y="sepal_length" ,palette="rainbow")
```

```python
sns.catplot(data=data,x="species",y="sepal_length",kind="box" ,palette="rainbow")
```

# Distribution plots
- histplot
- dist plot
- these two graphs are used  to check distribution of data

```python
sns.histplot(data=data,x="sepal_length",kde=True)
```

```python
sns.histplot(data=data,x="sepal_length",kde=True,hue="species")
```

```python
sns.distplot(x=data["sepal_length"])
```

# multivarient


```python
# pairplot
sns.pairplot(data,hue="species")
```

```python
data.corr()
```

```python
# heat map
sns.heatmap(data.corr(),annot=True)
```
